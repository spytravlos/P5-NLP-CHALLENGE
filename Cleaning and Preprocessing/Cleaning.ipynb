## First we will carry out the preprocessing of the data

```python
#make everything lowercase

df['user_review'] = df['user_review'].str.lower()
```

```python
df.head()
```

```python
#Remove special characters

df['user_review'] = df['user_review'].replace('[^a-zA-Z ]', '', regex = True)
```

```python
#Remove more than one space 
df['user_review'] = df['user_review'].replace('\s{2,}',' ',regex = True)
```

```python
df['user_review'][100]
```

```python
#Remove spaces at the beggining and at the end of the string

df['user_review'] = df['user_review'].replace(' $','',regex = True)
df['user_review'] = df['user_review'].replace('^ ','',regex = True)
```

```python
df['user_review'][100]
```

```python
df['user_review'][150]
```

### Remove stopwords


```python
from nltk.corpus import stopwords
stop = stopwords.words('english')

df['user_review'] = df['user_review'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)])) 
```

### Lemmatizing

```python
import nltk
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]
```

```python
df['user_review_cleaned'] = df['user_review_cleaned'].apply(lemmatize_text) 
```
